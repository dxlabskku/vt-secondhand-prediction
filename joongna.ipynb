```python
# 추가 인덱스 column drop

data = data.drop(data.columns[0], axis = 1)
data
```

```python
# title 분리

new_df = pd.DataFrame(columns=['기관', '선고 날짜', '판결문번호', '분류', '선고명', '공고번호'])

for i in range(len(data)):
    text = data.loc[i, 'tit']
    if '선고' in text:
        part1_and_2, part3= text.split(" 선고 ")
        part12 = part1_and_2.split(" ")
        part1 = part12[0]
        part2 = " ".join(part12[1:])
        new_df.loc[i, '기관'] = part1.strip()
        new_df.loc[i, '선고 날짜'] = part2.strip()
        new_df.loc[i, '판결문번호'] = part3.strip(" 판결")
    elif '자' in text:
        part1_and_2, part3= text.split("자")
        part12 = part1_and_2.split(" ")
        part1 = part12[0]
        part2 = " ".join(part12[1:])
        new_df.loc[i, '기관'] = part1.strip()
        new_df.loc[i, '선고 날짜'] = part2.strip()
        new_df.loc[i, '판결문번호'] = part3.strip(" 결정")
    else:
        new_df.loc[i, '기관'] = text.strip()
```

```python
new_df
```

```python
# sub_title 분리

for i in range(len(data)):
    text = data.loc[i, 'sub_tit']
    if '〈' in text:
        part1, part2_and_3 = text.split("〈")
        part2, part3 = part2_and_3.split("〉")
        new_df.loc[i, '분류'] = re.sub(r'[\[\]]', '', part1)
        new_df.loc[i, '선고명'] = part2.strip()
        new_df.loc[i, '공고번호'] = re.sub(r'[\[\]]', '', part3)
    elif '][' in text:
        part1, part2 = text.split("][")
        new_df.loc[i, '분류'] = re.sub(r'[\[\]]', '', part1)
        new_df.loc[i, '공고번호'] = re.sub(r'[\[\]]', '', part2)
    else:
        new_df.loc[i, '분류'] = text.strip()
```

```python
# 본문 분류 작업 전 [전문] 기준으로 나누기

new_df2 = pd.DataFrame(columns=['part1', 'part2'])

for i in range (len(data)):
    text = data.loc[i, 'info_1']
    part1, part2 = text.split('【전 문】')
    new_df2.loc[i, 'part1'] = part1.strip()
    new_df2.loc[i, 'part2'] = part2.strip()
```
